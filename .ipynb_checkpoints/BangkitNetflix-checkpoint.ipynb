```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('show_id', axis=1)
```

```python
dummies = pd.get_dummies(df['type'], drop_first=True)

df = pd.concat([df.drop('type',axis=1),dummies], axis=1)
```

```python
df.head()
```

```python
df['listed_in'].unique()
```

```python
df['listed_in'].nunique()
```

```python
df = df.drop('date_added', axis=1)
```

```python
df = df.drop('description', axis=1)
```

```python
df['TV Show Duration'] = df[df['TV Show'] == 0]['duration'].apply(lambda duration: int(duration[:-4]))
```

```python
df['Seasons'] = df[df['TV Show'] == 1]['duration'].map({'1 Season':'1', '2 Seasons':'2', '5 Seasons':'5',
                                        '3 Seasons':'3', '7 Seasons':'7',
                                        '4 Seasons':'4', '8 Seasons':'8', '6 Seasons':'6',
                                        '9 Seasons':'9', '14 Seasons':'14',
                                        '10 Seasons':'10', '12 Seasons':'12', '15 Seasons':'15',
                                        '11 Seasons':'11','13 Seasons':'13'})
```

```python
df['Duration or Seasons'] = df['TV Show Duration'].fillna(value=df['Seasons'])
```

```python
df.head()
```

```python
df = df.drop('TV Show Duration', axis=1)
df = df.drop('Seasons', axis=1)
```

```python
df = df.drop('duration', axis=1)
```

```python
df['Duration or Seasons'] = df['Duration or Seasons'].astype(float)
```

```python
df['first_genre'] = df['listed_in'].str.split(',', expand=True)[0]
```

```python
df['second_genre'] = df['listed_in'].str.split(',', expand=True)[1]
```

```python
df['third_genre'] = df['listed_in'].str.split(',', expand=True)[2]
```

```python
df['first_genre'].unique()
```

```python
df['first_genre'].nunique()
```

```python
dummies = pd.get_dummies(df['first_genre'], drop_first=True)

df = pd.concat([df.drop('first_genre',axis=1),dummies], axis=1)
```

```python
df[df['second_genre']=='Comedies']['second_genre']
```

```python
df['second_genre'] = df['second_genre'].astype(object)
```

```python
df['second_genre'] = df['second_genre'].str.replace(' ','')
```

```python
df['ComedyInt'] = np.where(df['second_genre']=='Comedies',1,0)
```

```python
for i in range(0,302):
    hasil = df[df['ComedyInt']==1].index[i]
    df['Comedies'].iloc[hasil] = 1
```
